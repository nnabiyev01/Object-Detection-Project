```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df = pd.read_csv("annotations_train.csv",header=None)
train_df
```

```python
def split_function(row):
    row = row.split("_")
    row = row[1]
    row = row.split(".")
    row = row[0]
    return row
def convert_train(row):
    row = str(row)
    opt_row = f"train_{row}.jpg"
    return opt_row
```

```python
train_df[0] = train_df[0].apply(split_function)
train_df[0] = train_df[0].astype(int)
train_df = train_df[train_df[0]<=200]
train_df[0] = train_df[0].apply(convert_train)
train_df.reset_index(inplace=True)
train_df = train_df.iloc[:,1:]
train_df.to_csv("train_ant.csv",index=None,header=None)
```
